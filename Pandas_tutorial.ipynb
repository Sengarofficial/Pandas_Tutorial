#### Difference between Nan values and null values and how to deal with it in pandas ? 

##### Null values represents "no value" or "nothing". It's not even an empty string or zero. it can be used to represent that nothing useful exists.
##### One possible way to handle null values is to remove them with: df.na.drop or one can change them to an actual value with df.na.fill(0)
##### Another way would be to select the rows where a specific column is null for further processing:
##### df.where(col('a').isNull()) and df.where(col("a").isNotNull())

##### NaN : Not a Number -> is a missing floating-point value, a special value that is part of the IEEE floating - point specification. 
##### type(np.nan) -> float , In numpy, the array witha NaN value is a native floating - point type array... np.array([1,2,3,np.nan]).dtype
##### However, when you try to do some arithmetic operations with NaN, the result will always be Nan : np.array([1,2,3,np.nan]).sum() -> nan, np.nan + 1 -> nan
##### Fortunately, Numpy provides some special aggregation methods that can ignore the existence of NaN value -> arr = np.array([1,2,3, np.nan]), np.nansum(arr) -> 6.0
##### Converting any NaN value into null : f = df.where(pandas.notnull(df), None)



#### None : A Python Object called None Type. 
#### np.array([1, 2, 3, None]) -> if you try to aggregate over this array, you will get an error because of the None Type. 
#### pandas is built to handle the two of them nearly intercheably, converting between them where appropriate. 
#### pd.series([1, np.nan, 2, None])
#### pandas automatically converts the None to NaN value.

#### As you can see this column has outliers(it is shown at boxplot) and it is right-skewed data(it is easily seen at histogram). Boxplot is the best way to see outliers..

##### Before handling outliers, we will detect them. We will use Tukey's rule to detect outliers. It is also known as the IQR rule.
##### First, we will calculate the Interquartile Range of the data (IQR = Q3 - Q1). Later, we will determine our outlier boundaries with IQR.
##### Later we will determine our outlier boundaries with IQR. we will get our lower boundary with this calculation Q1-1.5*IQR. 
##### We will get our upper boundary with this calculation Q3 + 1..5*IQR. According to this rule, the data between boundaries are acceptable but the data outside of the between lower and upper boundaries are outliers.
##### We can use 2.5 0r 2 to detect IQR. It depends on ouir data and analysis. But the most commonly used is 1.5 and we will use 1.5 IQR in this analysis.


```python
df[['table']].describe()
```

```python
sns.boxplot(x= df['table'])
plt.show()
```

##### With the describe method of pandas, we can see our data's Q1(25%) and Q3(75%) percentiles. We can calculate our IQR point and boundaries (with 1.5)

```python
df[['table']].describe()
```

```python
df[['table']].quantile(0.25)
```

```python
df[['table']].quantile(0.75)
```

```python
Q1 = df[['table']].quantile(0.25)
Q3 = df[['table']].quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_lim = Q1 - 1.5 * IQR
upper_lim = Q3 + 1.5 * IQR
```

```python
lower_lim
```

```python
upper_lim
```

##### Our upper boundary is 63.5 and our lower boundary is 51.5. This means that these values between 51.5 and 63.5 are acceprable but those outside mean there are ouliers. 
##### So we need to handle them because they coruupt our data. let's handle ouliers : 1. Dropping the outliers.
##### We can easily remove outliers , but this narrows our data. if we have lot of rows, big data, may be we can take risks.
##### But remember, if we drop the value, we delete all records (row). if we have vulnerable records, they can get lost.


```python
outliers_15_low = (df[['table']] < lower_lim)
outliers_15_up = (df[['table']] > upper_lim)
```

```python
len(df[['table']]) 
```

```python
(len(df[['table']][outliers_15_low]) + len(df[['table']][outliers_15_up]))
```

```python
(len(df[['table']][outliers_15_low]) + len(df[['table']][outliers_15_up])) - len(df[['table']])
```

```python
df[['table']][~(outliers_15_low | outliers_15_up) ]
```

#### 2. Winsorize Method : 

'''Our second method is the winsorize method. In this method, we limit outliers with an upper and lower limit. we will set the limits.
we will make our upper and lower limits for data our new maximum and minimum points. Let's check the boxplot again....... '''

```python
df = pd.read_csv(r'C:\Users\User\Desktop\diamonds.csv')
df.head()
```

```python
sns.boxplot(x = df['table'])
```

'''We have outliers, we detected them at the begining.For outliers, our upper limit is 63.5 and our lower limit is 51.5. For the Winsorize method, we have to import winsorize from scipy.
We need boundaries to apply winsorize. we will limit our data between 53 and 63. These values are within limits for outliers. We need precise points of these values in the percentile and we can use the quantile method of pandas.'''

```python
from scipy.stats.mstats import winsorize 
```

```python
upper_lim
```

```python
lower_lim
```

```python
df['table'].quantile(0.01)
```

```python
df['table'].quantile(0.98)
```

''' We will create a new variable with Winsorize Method. To implement the Winsorize Method, we write the exact boundary points as a tuple on the percentile.
For example, we will write(0.01, 0.02). This means we want tomapply quantile (0.01) and quantile (0.98) as a boundary. The first one is the exact point on percentile from the beginning, the second one is exact point on percentile from the end.'''


```python
df_table_win = winsorize(df['table'], (0.01, 0.02))
df_table_win
```

''' We applied Winsorize Method, let's check data on graphs ...'''

```python
sns.boxplot(x = df_table_win)
```

```python
sns.distplot(df_table_win, bins = 10, kde = False)
```

''' As you can see, there are no outliers. Focus on this, our maximum and minimum values on the boxplot; 53 and 63. We applied them as boundaries. Now, we can look at descriptive statistical values of old and new data with the describe method''' 

```python
df_table_win = pd.DataFrame(df_table_win)[0]
```

```python
df_table_win.describe()
```

```python
df['table'].describe()
```

''' Notice that, the mean and standard deviation of df_table_win have changed. Minimum and maximum points also changed but these changes do not affect median - logic descriptive statistical values.
Therefore, we should apply the winsorize method carefully because as you can see, mean - logic descriptive statistical values can change. It can corrupt our data, damage our analysis, or have adverse effects on models.'''
